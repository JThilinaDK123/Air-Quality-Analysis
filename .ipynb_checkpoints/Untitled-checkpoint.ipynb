## Load the datasets

```python
import pandas as pd
import requests
from io import StringIO
import matplotlib.pyplot as plt
import seaborn as sns
import ast
```

```python
# URL of the JSON file
json_url_indoor_air_quality = "https://fpardrecruiting002st.z33.web.core.windows.net/CodeEvaluation/DataScientist/iaq.json"
json_url_outdoor_air_quality = "https://fpardrecruiting002st.z33.web.core.windows.net/CodeEvaluation/DataScientist/oaq.json"
json_url_sensor_air_quality = "https://fpardrecruiting002st.z33.web.core.windows.net/CodeEvaluation/DataScientist/floors.json"
```

```python
# Fetch the JSON content from the URL
response_1 = requests.get(json_url_indoor_air_quality)
response_2 = requests.get(json_url_outdoor_air_quality)
response_3 = requests.get(json_url_sensor_air_quality)
json_data_url_indoor_air_quality = response_1.json()
json_data_url_outdoor_air_quality = response_2.json()
json_data_url_sensor_air_quality = response_3.json()

# Convert the JSON data to a DataFrame
df_json_url_indoor_air_quality = pd.DataFrame(json_data_url_indoor_air_quality)
df_json_url_outdoor_air_quality = pd.DataFrame(json_data_url_outdoor_air_quality)
df_json_url_sensor_air_quality = pd.DataFrame(json_data_url_sensor_air_quality)
```

```python
# Convert string representation of dictionaries and lists to actual dictionaries and lists
df_floor_data["image"] = df_floor_data["image"].apply(ast.literal_eval)
df_floor_data["sensors"] = df_floor_data["sensors"].apply(ast.literal_eval)

# Create a new DataFrame to store the result
df_floor_result_df = pd.DataFrame()

# Iterate through rows and explode sensors column
for index, row in df_floor_data.iterrows():
    sensors_df = pd.json_normalize(row["sensors"])
    sensors_df["id"] = row["id"]
    sensors_df["image"] = row["image"]["name"]
    df_floor_result_df = pd.concat([df_floor_result_df, sensors_df], ignore_index=True)

# Rename columns for clarity
df_floor_result_df.rename(columns={"sensor_id": "sensor_id", "location.y": "y", "location.x": "x"}, inplace=True)

# Display the result
df_floor_result_df
```

## Merge datasets

#### Indoor Air Quality + Floor Plans

```python
# Merge indoor air quality data with floor data
merged_data = pd.merge(df_json_url_indoor_air_quality, 
                       df_floor_result_df, on='sensor_id', how='left')
merged_data
```

```python
# Visualize indoor temperature distribution
plt.figure(figsize=(12, 6))
sns.boxplot(x='id', y='temp', data=merged_data)
plt.title('Indoor Temperature Distribution Across Rooms and Floors')
plt.show()

# Visualize indoor humidity distribution
plt.figure(figsize=(12, 6))
sns.boxplot(x='id', y='humidity', data=merged_data)
plt.title('Indoor Humidity Distribution Across Rooms and Floors')
plt.show()

# Visualize indoor CO2 level distribution
plt.figure(figsize=(12, 6))
sns.boxplot(x='id', y='co2', data=merged_data)
plt.title('Indoor CO2 Level Distribution Across Rooms and Floors')
plt.show()
```

#### Trend Analysis for indoor air quality

#### 01) Temperature

```python
# Convert 'documentTime' to datetime format
merged_data['documentTime'] = pd.to_datetime(merged_data['documentTime'])

# Extract month and hour for seasonal analysis
merged_data['month'] = merged_data['documentTime'].dt.month
merged_data['hour'] = merged_data['documentTime'].dt.hour

# Filter data for winter months (assuming December, January, February)
winter_data = merged_data[(merged_data['month'] >= 12) | (merged_data['month'] <= 2)]

# Visualize seasonal trends
plt.figure(figsize=(12, 6))
sns.lineplot(x='hour', y='temp', data = winter_data, hue='id')
plt.title('Seasonal Trends in Indoor Temperature During Winter Months')
plt.show()
```

#### 02) Humidity

```python
# Visualize seasonal trends
plt.figure(figsize=(12, 6))
sns.lineplot(x='hour', y='humidity', data = winter_data, hue='id')
plt.title('Seasonal Trends in Indoor Humidity During Winter Months')
plt.show()
```

#### 03) Co2

```python
# Visualize seasonal trends
plt.figure(figsize=(12, 6))
sns.lineplot(x='hour', y='co2', data = winter_data, hue='id')
plt.title('Seasonal Trends in Indoor Co2 During Winter Months')
plt.show()
```

# Define operation goals

```python
temperature_goal = 21
humidity_goal = 20
co2_goal = 1000

# Filter data for work hours 
work_hours_data = merged_data[(merged_data['hour'] >= 9) & (merged_data['hour'] <= 17)]

# Check compliance for temperature
temperature_compliance = work_hours_data['temp'] >= temperature_goal

# Check compliance for humidity
humidity_compliance = work_hours_data['humidity'] >= humidity_goal

# Check compliance for CO2 levels
co2_compliance = work_hours_data['co2'] <= co2_goal

# Display compliance summary
compliance_summary = pd.DataFrame({
    'Temperature': temperature_compliance,
    'Humidity': humidity_compliance,
    'CO2': co2_compliance
})

print("Compliance Summary during Work Hours:")
print(compliance_summary.describe())
```

1. **Count:**
   - Temperature count: 38,195
   - Humidity count: 38,195
   - CO2 count: 38,195

   This indicates that there are 38,195 data points or observations for each parameter during work hours.

2. **Unique:**
   - Temperature unique values: 2
   - Humidity unique values: 1
   - CO2 unique values: 2

   This suggests that there are two unique values for Temperature and CO2, and only one unique value for Humidity. This implies that Temperature and CO2 may have two possible states (True and False), while Humidity seems to have only one (True). "True" means the operational goal is success in working hours and "False" means the operational goal is not success in working hours

3. **Top:**
   - Top Temperature value: True
   - Top Humidity value: True
   - Top CO2 value: True

   The "Top" value represents the most frequently occurring value in the dataset. In this case, the most common value for Temperature, Humidity, and CO2 is True.

4. **Frequency (Freq):**
   - Frequency of top Temperature value: 37,459
   - Frequency of top Humidity value: 38,195
   - Frequency of top CO2 value: 36,929

   This indicates how often the most common value occurs in the dataset. For example, True occurs 37,459 times for Temperature, 38,195 times for Humidity, and 36,929 times for CO2.

In summary, during work hours:
- Temperature and CO2 have two possible states (True and presumably False).
- Humidity has only one unique state (True).
- True is the most common value for all three parameters, occurring very frequently in the dataset.
- This means, in most of the time frames, the operational goals in working hours getting success
